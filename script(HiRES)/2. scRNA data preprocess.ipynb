```python
import anndata
import pandas as pd
```

```python
scRNA_metadata = pd.read_csv('/lmh_data/data/sclab/GSE223917/GSE223917_HiRES_emb_metadata.csv')
scRNA_metadata
```

```python
# target_label = ['Astro', 'Endo', 'Oligo', 'OPC', 'Pvalb', 'Sst', 'Vip']
# target_scRNA_metadata = scRNA_metadata[scRNA_metadata['subclass_label'].isin(target_label)]
# target_scRNA_metadata

target_scRNA_metadata = scRNA_metadata.copy()
target_scRNA_metadata = target_scRNA_metadata.set_index('Cellname')
target_scRNA_metadata.index.name = 'sample_name'
target_scRNA_metadata
```

```python
infos = pd.DataFrame()
for chunk in pd.read_table('/lmh_data/data/sclab/GSE223917/GSE223917_HiRES_emb.rna.umicount.tsv', chunksize=10000):
    infos = pd.concat([infos, chunk])

infos = infos.set_index('gene')
infos
```

```python
infos = infos.T
infos.index.name = 'sample_name'
infos
```

```python
infos = anndata.AnnData(X=infos)
infos
```

```python
infos.obs
```

```python
_target_scRNA_metadata = target_scRNA_metadata.copy()
_target_scRNA_metadata = _target_scRNA_metadata.loc[infos.obs.index]
infos.obs["cell_type"] = _target_scRNA_metadata['Celltype']
infos.obs['domain'] = 'scRNA'
infos.obs
```

```python
infos.write("/lmh_data/data/sclab/GSE223917/scRNA.h5ad", compression="gzip")
```
