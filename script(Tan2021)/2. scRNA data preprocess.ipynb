```python
import anndata
import pandas as pd
```

```python
scRNA_metadata = pd.read_csv('/lmh_data/data/sclab/Mouse_10x/metadata.csv')
scRNA_metadata
```

```python
scRNA_metadata['cluster_label'] = scRNA_metadata['cluster_label'].str.split('_').str[-1]
scRNA_metadata['cluster_label'].unique()
```

```python
target_label = ['Astro', 'Micro-PVM', 'Oligo']
target_scRNA_metadata = scRNA_metadata[scRNA_metadata['cluster_label'].isin(target_label)]
target_scRNA_metadata
```

```python
import h5py
import numpy as np

infos = h5py.File('/lmh_data/data/sclab/Mouse_10x/expression_matrix.hdf5')
infos['data'].keys()
```

```python
_samples = np.array(infos['data']['samples']).astype(str)
_samples_id = []
for sample_name in target_scRNA_metadata['sample_name']:
    _samples_id.append(np.where(_samples==sample_name)[0][0])
_samples_id.sort()
infos['data']['counts'][:, _samples_id].shape
```

```python
infos = anndata.AnnData(
    X = infos['data']['counts'][:, _samples_id],
    obs = np.array(infos['data']['gene']).astype(str),
    var = np.array(infos['data']['samples'][_samples_id]).astype(str),
    dtype = infos['data']['counts'].dtype.name,
)
infos
```

```python
infos.var.columns, infos.obs.columns = ['sample_name'], ['gene_name']

infos.var = infos.var.set_index('sample_name')
infos.obs = infos.obs.set_index('gene_name')
infos = infos.T
infos
```

```python
infos.obs
```

```python
_target_scRNA_metadata = target_scRNA_metadata.set_index('sample_name')
_target_scRNA_metadata = _target_scRNA_metadata.loc[infos.obs.index]
infos.obs["cell_type"] = _target_scRNA_metadata['cluster_label']
infos.obs['domain'] = 'scRNA'
infos.obs
```

```python
infos.write("/lmh_data/data/sclab/Mouse_10x/scRNA.h5ad", compression="gzip")
```
