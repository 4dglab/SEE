```python
import os
import sys
sys.path.append('/lmh_data/work/SEE/analyse/')

import anndata
import numpy as np
import pandas as pd
import scanpy as sc

from matplotlib.pyplot import rc_context
import matplotlib.pyplot as plt
import seaborn as sns

from analyse_util import figure_size, set_plt, set_Border
```

```python
rna = anndata.read_h5ad("/lmh_data/data/sclab/sclab/rna_result.h5ad")
hic = anndata.read_h5ad("/lmh_data/data/sclab/sclab/hic_result.h5ad")
```

```python
combined = anndata.concat([rna, hic])
sc.pp.neighbors(combined, use_rep="X_glue", metric="cosine")
sc.tl.umap(combined)
```

```python
# set_plt(figsize=(5, 5))
# sns.set_theme(style="whitegrid")

# fig, ax = plt.subplots()
with rc_context({'figure.figsize': (5, 5)}):
    sc.pl.umap(combined, color=["domain"], add_outline=True, legend_loc=None, title='', frameon=False, show=False)
    plt.savefig('/lmh_data/work/SEE/analyse/figure_file/mapping.pdf', format='pdf', bbox_inches='tight')
```
