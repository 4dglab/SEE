```python
import pandas as pd
```

```python
v19_anno = pd.read_csv('/lmh_data/data/sclab/public/gencode.v19.annotation.gtf',
                       header=None, sep='\t', skiprows=[i for i in range(5)], usecols=[0, 2, 3, 4, 6, 8], names=[
                           'chrom', 'type', 'start', 'end', 'strand', 'info'
                       ])
v19_anno['gene_name'] = v19_anno['info'].str.extract(r'(gene_name ")(\w*)')[1]
v19_anno['transcript_name'] = v19_anno['info'].str.extract(r'(transcript_name ")(\S*)(")')[1]
v19_anno
```

```python
v19_anno['transcript_type'] = v19_anno['info'].str.extract(r'(transcript_type ")(\S*)(")')[1]
v19_anno
```

```python
v19_anno['transcript_type'].unique()
```

```python
def test(genes):
    filter_genes_info = v19_anno[
        (v19_anno['transcript_type']=='protein_coding')&
        (v19_anno['gene_name'].isin(genes))&
        (v19_anno['type']=='gene')
    ]
    print(filter_genes_info.shape)
```
