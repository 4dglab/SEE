```python
import anndata
import pandas as pd
```

```python
scRNA_metadata = pd.read_csv('/home/micl/workspace/lmh_data/Human_M1_10x/metadata.csv')
scRNA_metadata
```

```python
scRNA_metadata['subclass_label'].unique()
```

```python
target_label = ['Astro', 'Endo', 'Oligo', 'OPC', 'Pvalb', 'Sst', 'Vip']
target_scRNA_metadata = scRNA_metadata[scRNA_metadata['subclass_label'].isin(target_label)]
target_scRNA_metadata
```

```python
infos = pd.DataFrame()
for chunk in pd.read_csv('/home/micl/workspace/lmh_data/Human_M1_10x/matrix.csv', chunksize=10000):
    _filter = chunk[chunk['sample_name'].isin(target_scRNA_metadata['sample_name'])]
    infos = pd.concat([infos, _filter])

_filter = None  # Prevent memory hogging
infos
```

```python
infos = infos.set_index('sample_name')
infos = anndata.AnnData(X=infos)
infos
```

```python
infos.obs
```

```python
_target_scRNA_metadata = target_scRNA_metadata.set_index('sample_name')
_target_scRNA_metadata = _target_scRNA_metadata.loc[infos.obs.index]
infos.obs["cell_type"] = _target_scRNA_metadata['subclass_label']
infos.obs['domain'] = 'scRNA'
infos.obs
```

```python
infos.write("/home/micl/workspace/lmh_data/Human_M1_10x/scRNA.h5ad", compression="gzip")
```
